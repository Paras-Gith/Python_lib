```python
Years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
orange = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896]
```

```python
#ploting multiple line using plt.plot
plt.plot(Years, apples)
plt.plot(Years, orange)
plt.xlabel('Years')
plt.ylabel('yeild(tons per hectare)');
```

```python
#chart title and legend using plt.legend and plt.title, these are using for naming line in the graph
plt.plot(Years, apples)
plt.plot(Years, orange)

plt.xlabel('Years')
plt.ylabel('yeild(tons per hectare)')

plt.title("crop yeild in kanto")
plt.legend(['Apples', 'Orange']);
```

```python
#line marker For showing data points using marker and there are lot of marker in matplot
plt.plot(Years, apples, marker = 'o')
plt.plot(Years, orange, marker = 'x')

plt.xlabel('Years')
plt.ylabel('yeild(tons per hectare)')

plt.title("crop yeild in kanto")
plt.legend(['Apples', 'Orange']);
```

```python
#styling line and marker 
# color or C : set the color of the line
# linesytle or ls : choose between a solid or dashed line
# lineweidth or lw: set the width of a line
# makersize or ms : set the size of the marker
# markeredgecolor or mec : set the edge color for markers
# markeredgewidth or mew : set the edge width for markers                                    
# markeracecolor or mfc : set the fill color for markers
# aplha : opacity of the plot                               
```

```python
plt.plot(Years, apples, marker = 's', c='b', ls='-', lw=2, ms= 8, mew = 2, mec = 'navy')
plt.plot(Years, orange, marker = 'o', c='r', ls='--', lw=3, ms=10, alpha=.5 )

plt.xlabel('Years')
plt.ylabel('yeild(tons per hectare)')

plt.title("crop yeild in kanto")
plt.legend(['Apples', 'Orange']);
```

```python
#for shorthand for specifing line style using fmt or fmt = '[marker][line][color]'
plt.plot(Years, apples, 's-b')
plt.plot(Years, orange, 'o--r')

plt.xlabel('Years')
plt.ylabel('yeild(tons per hectare)')

plt.title("crop yeild in kanto")
plt.legend(['Apples', 'Orange']);
```

```python
#changing the figure size using .figure
plt.figure(figsize=(8,4))

plt.plot(Years, apples, 's-b')
plt.plot(Years, orange, 'o--r');
```

```python
#load data into a pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

```python
#adding hues using .hue
sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100);
```

```python
#customizing seaborn figure
plt.figure(figsize =(12, 6))
plt.title('space Dimensions')

sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100);
```

```python
#plotting using pandas data frame
plt.title('space Dimensions')

sns.scatterplot(x='sepal_length',
                y='sepal_width',
               hue=flowers_df.species,
               s=100,
               data=flowers_df);
```

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

```python
#controlling the size and number of bins using bins aggrument(bin is statical term)
plt.hist(flowers_df.sepal_width, bins=5);
```

```python
import numpy as np
```

```python
np.arange(2, 5, 0.25) #it means that make a range from 2 to 5 with a gap of 0.25
```

```python
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25 ));
```

```python
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

```python
#..........................multiple histogram.........................
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df =flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.legend(['setosa', 'versicolor']);
```

```python
plt.title('Distribution of sepal width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, versicolor_df.sepal_width], 
         bins=np.arange(2, 5, 0.25),
        stacked=True);
plt.legend(['setosa', 'versicolor', 'virginica']);
```

```python
#bar plots with averages using dataset called tips
tips_df = sns.load_dataset("tips");
```

```python
bill_avg_df = tips_df.groupby('day', observed=False)[['total_bill']].mean()
bill_avg_df
```

```python
sns.barplot(x='day', y='total_bill', data=tips_df);
#line between the bar are variation
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df);
```
